```python
import numpy as np
import matplotlib.pyplot as plt

from k_means_clustering import KMeansClustering
```

```python
cluster_1 = np.random.randn(100, 2) + np.array([0, 0])
cluster_2 = np.random.randn(100, 2) + np.array([5, 5])
cluster_3 = np.random.randn(100, 2) + np.array([0, 5])

X = np.vstack((cluster_1, cluster_2, cluster_3))
```

```python
model = KMeansClustering(k=3)
model.fit(X)
```

```python
labels = model.predict(X)
centroids = np.array(model.centroids)

print("Centroids:")
print(centroids)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', c='r', s=50)
plt.title("KMeans Clustering Result")
plt.show()
```

```python
best_k, inertias = model._find_best_k(X)

print("Best k:", best_k)
print("Inertias:", inertias)

k_values = range(1, len(inertias) + 1)

plt.figure()
plt.plot(k_values, inertias, marker='o')
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
