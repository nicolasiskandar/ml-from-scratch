```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

from linear_regression import LinearRegression
```

```python
n_samples = 1000
n_features = 10
X, y = make_regression(n_samples=n_samples, n_features=n_features, n_informative=5, noise=10, random_state=42)
```

```python
X_fit, X_val, y_fit, y_val = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], y, color="purple", marker="o", s=4)
plt.title("Feature vs Target Distribution")
plt.xlabel("Feature 0")
plt.ylabel("Target")
plt.grid(alpha=0.1)
plt.show()
```

```python
model = LinearRegression(learning_rate=0.01)
model.fit(X_fit, y_fit)
```

```python
y_pred_train = model.predict(X_fit)

fig = plt.figure(figsize=(6,4))
plt.scatter(y_fit, y_pred_train, color="green", s=4)
plt.plot([min(y), max(y)], [min(y), max(y)], color="lightblue", linewidth=2)
plt.xlabel("True training values")
plt.ylabel("Predicted training values")
plt.show()
```

```python
y_pred_val = model.predict(X_val)

fig = plt.figure(figsize=(6,4))
plt.scatter(y_val, y_pred_val, color="magenta", s=4)
plt.plot([min(y), max(y)], [min(y), max(y)], color="lightblue", linewidth=2)
plt.xlabel("True validation values")
plt.ylabel("Predicted validation values")
plt.show()
```

### Using L2 Regularization

```python
model = LinearRegression(learning_rate=0.01, lambda_=0.1)
model.fit(X_fit, y_fit)
```

```python
y_pred_train = model.predict(X_fit)

fig = plt.figure(figsize=(6,4))
plt.scatter(y_fit, y_pred_train, color="green", s=4)
plt.plot([min(y), max(y)], [min(y), max(y)], color="lightblue", linewidth=2)
plt.xlabel("True values")
plt.ylabel("Predicted train values")
plt.show()
```

```python
y_pred_val = model.predict(X_val)

fig = plt.figure(figsize=(6,4))
plt.scatter(y_val, y_pred_val, color="magenta", s=5)
plt.plot([min(y), max(y)], [min(y), max(y)], color="lightblue", linewidth=2)
plt.xlabel("True values")
plt.ylabel("Predicted validation values")
plt.show()
```
